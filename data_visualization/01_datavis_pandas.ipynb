```python
# Import libararies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df = sns.load_dataset('iris')
df.head()
```

## _**1. Plot with One Variable:**_

### _**1.1 Histogram**_

- Univariate Analysis
- Numerical col
- Use case - Frequency Count

```python
# Method 1. To plot all the numeric variables in a dataset
df.hist()
plt.show()
```

```python
# Method 2. To plot all the numeric variables in a dataset
df.plot.hist()
plt.show()
```

```python
# Method 3. To plot all the numeric variables in a dataset
df.plot(kind = 'hist')
plt.show()
```

```python
# Method 1. To plot only one variable 
df['sepal_width'].plot(kind = 'hist')
plt.show()
```

```python
# Method 2. To plot only one variable 
df.plot(y = 'sepal_width', kind = 'hist')
plt.show()
```

```python
# Method 1: To plot two or more variable 
df.plot(y = ['sepal_length', 'petal_length'], kind = 'hist')
plt.show()
```

```python
# Method 2: To plot two or more variable 
df[['sepal_length', 'petal_length']].plot(kind = 'hist')
plt.show()
```

```python
df.plot.hist(y='sepal_length')
plt.show()
```

### _**1.2 Boxplot**_

- Univariate Analysis
- Numerical col
- Use case - Outlier Detection

```python
# Method 1: To plot all the numeric variables in a DataFrame
df.boxplot()
plt.show()
```

```python
# Method 2: To plot all the numeric variables in a DataFrame
df.plot(kind = 'box')
plt.show()
```

```python
# Method 1: To plot one numeric variable in a DataFrame
df.plot(y = 'sepal_length', kind = 'box')
plt.show()
```

```python
# Method 2: To plot one numeric variable in a DataFrame
df['sepal_length'].plot(kind = 'box')
plt.show()
```

```python
# Method 1: To plot two or more numeric variables in a DataFrame
df.plot(y=['sepal_length', 'sepal_width'], kind = 'box')
plt.show()
```

```python
# Method 2: To plot two or more numeric variables in a DataFrame
df[['sepal_length', 'sepal_width']].plot(kind = 'box')
plt.show()
```

```python
df.plot.box(y='sepal_length')
```

## _**2. Plot with Two Variables:**_

### _**2.1 Scatter Plot:**_

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# Method 1: To plot scatter plot we need to give two of numeric variables
df.plot(x='sepal_length', y='sepal_width', kind = 'scatter')
plt.show()
```

```python
# Method 2: To plot scatter plot we need to give two of numeric variables
df.plot.scatter(x='sepal_length', y='sepal_width', color='red')
plt.show()
```

### _**2.2 Bar Chart/Plot:**_

- Bivariate Analysis
- Categorical vs Numerical(aggregate function)
- Use case - Aggregate analysis of groups

```python
# Method 1: To plot bar plot we need to give one categorical and other numeric variable aggregate value(mean,sum, ect.)
df.groupby('species')['petal_length'].mean().plot(kind = 'bar', legend= True)
plt.show()
```

```python
# Method 1: To plot one categorical and two numeric variable aggregate value(mean,sum, ect.)
df.groupby('species')[['petal_length', 'petal_width']].mean().plot(kind = 'bar', legend= True)
plt.show()
```

```python
# Method 1: To plot horizental box plot, we need to give kind='barh'
df.groupby('species')[['petal_length', 'petal_width']].mean().plot(kind = 'barh', legend= True)
plt.show()
```

### _**2.3 Pie Chart/Plot:**_

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# Method 1: To plot pie plot we need to give one categorical and other numeric variable aggregate value(mean,sum, ect.)
df.groupby('species')['petal_length'].mean().plot(kind = 'pie')
plt.show()
```

## _**Summary**_

```python
# Univariate Analysis (One variable)
df['petal_length'].plot(kind='hist') # to plot histogram
df['petal_length'].plot(kind='box')  # to plot box plot

# Bivairate Analysis (two variables (one categorical and other numerical (aggregate)))
df.groupby('species')['petal_length'].mean().plot(kind='bar')   # to plot bar plot
df.groupby('species')['petal_length'].mean().plot(kind='barh')  # to plot horiziental bar plot
df.groupby('species')['petal_length'].mean().plot(kind='pie')   # to plot pie plot

# Bivairate Analysis (two numerical variables)
df.plot(kind = 'scatter', x='petal_length', y='sepal_length')   # to plot scatter plot
```
