```python
# Import libararies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df = sns.load_dataset('iris')
df.head()
```

## _**1. Plot with One Variable:**_

### _**1.1 Histogram**_

- Univariate Analysis
- Numerical col
- Use case - Frequency Count

```python
# to disable the warnings
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Method 1. To plot histogram of one numeric variable in a dataset
sns.histplot(data=df, x='petal_length')
plt.show()
```

```python
# Method 1. To plot histogram of one numeric variable in a dataset
sns.histplot(data=df, x='petal_length', hue='species')
```

### _**1.2 Boxplot**_

- Univariate Analysis
- Numerical col
- Use case - Outlier Detection

```python
# Method 1: To plot one numeric variables in a DataFrame
sns.boxplot(data = df, x='petal_length')
```

```python
sns.boxplot(data = df)
```

## _**2. Plot with Two Variables:**_

### _**2.1 Scatter Plot:**_

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# Method 1: To plot scatter plot we need to give two of numeric variables
sns.scatterplot(data = df, x='sepal_length', y='petal_length')
```

```python
# we can use hue
sns.scatterplot(data = df, x='sepal_length', y='petal_length', hue='species')
```

### _**2.2 Bar Chart/Plot:**_

- Bivariate Analysis
- Categorical vs Numerical(aggregate function)
- Use case - Aggregate analysis of groups

```python
# to print al the 
sns.barplot(data = df)
```

```python
# to plont only one
sns.barplot(data = df, x='petal_length')
```

### _**2.3 Heatmap:**_

- Multivariate analysis
- Numeric data
- use Case: Usually used for Correlation Analysis

```python
# for heat map firs we need to find the correlation of numeric values
sns.heatmap(df.iloc[:,0:4].corr(), linewidths=0.1, annot=True)
```
