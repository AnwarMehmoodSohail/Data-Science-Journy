```python
# Import libararies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df = sns.load_dataset('iris')
df.head()
```

## _**1. Plot with One Variable:**_

### _**1.1 Histogram**_

- Univariate Analysis
- Numerical col
- Use case - Frequency Count

```python
# Method 1. To plot histogram of one numeric variable in a dataset
plt.hist(df['petal_length'])
plt.title("Histogram of Petal Length")
plt.ylabel("Frequency Count")
plt.show()
```

### _**1.2 Boxplot**_

- Univariate Analysis
- Numerical col
- Use case - Outlier Detection

```python
# Method 1: To plot one numeric variables in a DataFrame
plt.boxplot(df['petal_length'])
plt.xlabel('Petal Length')
plt.show()
```

```python
# Method 1: To plot two or more numeric variables in a DataFrame
plt.boxplot(df[['petal_length', 'petal_width']])
plt.show()
```

## _**2. Plot with Two Variables:**_

### _**2.1 Scatter Plot:**_

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# Method 1: To plot scatter plot we need to give two of numeric variables
plt.scatter(df['petal_length'], df['petal_width'], color='red', marker='D')
plt.show()
```

### _**2.2 Bar Chart/Plot:**_

- Bivariate Analysis
- Categorical vs Numerical(aggregate function)
- Use case - Aggregate analysis of groups

```python
# we need to give species as categorical variavle and mean() of petal_length as aggregate function
plt.bar(df['species'], df['petal_length'].mean())
plt.show()
```

### _**2.3 Pie Chart/Plot:**_

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# we need to give 'petal length' as a 'numeric values' and names of 'species' as 'categorical values' as label
plt.pie(df.groupby('species')['petal_length'].mean(), labels=df['species'].unique())
plt.show()
```

```python
# To add percentage used 'autopct' and to seperate one part use 'explode'
plt.pie(df.groupby('species')['petal_length'].mean(), labels=df['species'].unique(), autopct='%0.1f%%', explode=[0.1, 0, 0])
plt.show()
```

## _**Summary**_

```python
# Univariate Analysis (One variable)
plt.hist(df['petal_length']) # to plot histogram
plt.boxplot(df['petal_length'])  # to plot box plot

# # Bivairate Analysis (two variables (one categorical and other numerical (aggregate)))
plt.bar(df['species'], df['petal_length'].mean())   # to plot bar plot
plt.barh(df['species'], df['petal_length'].mean())   # to plot horiziental bar plot
plt.pie(df.groupby('species')['petal_length'].mean(), labels=df['species'].unique())   # to plot pie plot

# Bivairate Analysis (two numerical variables)
plt.scatter(df['petal_length'], df['petal_width'])  # to plot scatter plot
```
