# _**2. Catergorical Distribution Plot**_
## _**2.1 Boxplot**_

A boxplot, also known as a whisker plot, is a graphical representation of statistical data based on a five-number summary. These five numbers are minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

In the boxplot:

- The box represents the interquartile range (IQR), which is the range between the first quartile (25th percentile) and the third quartile (75th percentile).
- The line inside the box represents the median (50th percentile) of the data.
- The whiskers represent the data within 1.5 times the IQR. Any data points outside this range are considered outliers and are represented as individual points.

Boxplots are useful for comparing distributions across different categories.

```python
# axes level function
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
# axes level function
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
# for single variable we need to give on y-axis and numeric variable
sns.boxplot(data=tips, y='total_bill')
```

```python
# function level function
sns.catplot(data=tips, kind='box', x='day', y='total_bill', hue='sex')
```

# _**2.3 Bar Plot:**_

A bar plot or bar chart is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. 

A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value (mean/median etc). 

In the context of seaborn, a bar plot is a plot that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The `barplot` function in seaborn draws a bar plot from a DataFrame.


```python
# axes level funciton
sns.barplot(data=tips, x='day', y='total_bill')
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars, as we can see in the above plot

```python
# to calculate the minimum value
sns.barplot(data=tips, x='day', y='total_bill', estimator=np.min)
```

```python
# to calculate the maximum value
sns.barplot(data=tips, x='day', y='total_bill', estimator=np.max)
```

```python
# to remove erro bar we can use ci=None
sns.barplot(data=tips, x='sex', y='total_bill',ci=None)
```

# _**2.3 Count Plot:**_

A count plot can be thought of as a histogram across a categorical variable, instead of a quantitative one. It's a type of bar plot where the y-axis represents the count of the number of occurrences of each category in the dataset.

In the context of seaborn, a count plot can be created using the `countplot()` function. It's often used to visualize the distribution of categorical data.

Here's a basic example of how to use it:



```python
sns.countplot(data=tips, x='day', hue='sex')
```

```python
# figure level funtion
sns.catplot(data=tips, kind='count', x='sex', col='day', row='smoker')
```

# _**3 Multiplot:**_

## _**3.1 Pair Plot:**_

A pairplot is a matrix of scatterplots that allows you to visualize the pairwise relationships between different (numerical) variables in a dataset. It's a quick way to examine the relationships between every pair of variables in your data.

In the context of seaborn, a pairplot can be created using the `pairplot()` function. It creates a grid of Axes such that each variable in your data will be shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are usually filled with histograms or KDE plots.

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris, hue='species')
```

## _**Facit Grid:**_

`FacetGrid` is a class in seaborn that allows you to create a grid of subplots based on a feature (or features) in your dataset. It's a great way to create multiple plots that are conditioned on certain variable values.

Here's a basic example of how to use it:

## _**3.2 Pair Grid:**_

`PairGrid` is a subplot grid for plotting pairwise relationships in a dataset. It is a class in seaborn which is used to create a grid of subplots using the same plot type to visualize data.

Unlike `pairplot`, `PairGrid` is a more flexible level of interface, allowing you to control the types of plots in each cell of the grid. It allows you to quickly draw a grid of small subplots using the same plot type to visualize data in each. This can be very helpful for exploring correlations between multidimensional data, when you'd like to plot all pairs of values against each other.

Here's a basic example of how to use it:



```python
#  # To plot all scatter plot
g = sns.PairGrid(data=iris)
g.map(sns.scatterplot)
```

```python
 # To plot diagnal differene and off diagnol differen
g = sns.PairGrid(data=iris)
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
 # To plot diagnal differene and off diagnol differen
g = sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

# _**3.3 Joint Plot:**_

`jointplot` is a function in seaborn that allows you to essentially create a scatterplot with two additional histograms, one for each variable (x and y in the scatterplot). This can be very useful when you want to visualize the relationship between two variables, along with their individual distributions.

By default, `jointplot` creates a scatterplot in the center, a histogram in the top margin, and a histogram in the right margin.


```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
# can change the kind
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

```python
# can change the kind
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hist')
```

```python
# can change the kind
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

# _**3.4 Joint Grid:**_

`JointGrid` is a class in seaborn for plotting a bivariate plot with marginal univariate plots. It's a grid layout that forms the basis for other seaborn functions like `jointplot`.

The `JointGrid` class allows for a more flexible customization of the plots. It creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.

In this example, a scatter plot is used for the joint relationship between 'x' and 'y', and a histogram is used for the marginal distributions.

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.histplot)
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip', hue='sex')
g.plot(sns.scatterplot, sns.kdeplot)
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip', hue='sex')
g.plot(sns.histplot, sns.violinplot)
```

# _**4. Matrix Plot:**_

## _**4.1 Heatmap:**_

A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colors. It's a way of visualizing a dataset and is particularly useful when you want to show variance across multiple variables, revealing any patterns, displaying whether any variables are similar to each other, and for detecting if any correlations exist in-between them.

In seaborn, a heatmap can be created using the `heatmap()` function. It uses a color-coded matrix to represent the data.


```python
# import data
import plotly.express as px
gap = px.data.gapminder()
gap
```

```python
# We can heatmap to check the life expectency of Europe countries year wise
temp_gap = gap[gap['continent']=='Europe']
temp_gap = temp_gap.pivot(index='country', columns='year', values='lifeExp')
temp_gap
```

```python
# to plot the above pivot using heatmap
plt.figure(figsize=(10,10))
sns.heatmap(data=temp_gap, annot=True, linewidths=1)
```

```python
# correlation matrix:  we can also use plot the correlation matrix using heatmap
temp_iris = iris.loc[:, ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
temp_iris.corr()  # find the correlation
```

```python
# plot correlation using heatmap
sns.heatmap(temp_iris.corr(), cmap='summer', annot=True)
```

## _**4.2 Clustermap:**_

A `clustermap` is a type of plot that is a specific implementation of a heatmap. It not only visualizes the data in a matrix format, but also uses hierarchical clustering (both row-wise and column-wise) to group similar rows and columns together. This can make patterns more visible.

In seaborn, a clustermap can be created using the `clustermap()` function. It uses a color-coded matrix to represent the data and dendrograms along the axes to represent the clustering.

Here's a basic example of how to use it:

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
